```python
import matplotlib.pyplot as plt
```

```python
n=6
bits=[]
nn_states=[]
as_states=[]
for i in range(1,n):
    bits.append(i)
    nn_states.append(2**i)
    as_states.append(i)


plt.plot(bits,nn_states,c="green")
plt.plot(bits,as_states,c="red")
plt.xlabel("No. Of Bits")
plt.ylabel("No. Of States")
plt.title("Nearest Neighbour vs Adder Subtractor\nGreen=NN | Red=Adder Subtractor")
plt.show()
print("We Get Exponential(O(2^N)) Scale up In Nearest Neighbour Approach Compare to Adder Subtractor Approach For N Qubits")
```

```python
def nn_gate_count_cal(n):
    step=(2**(n-1))-1
    s=0
    for i in range(step):
        if i%2 == 0:
            s+=5
        else:
            s+=7
    return s

s=6
states=[]
nn_gate_count=[]
as_gate_count=[]
for i in range(1,s):
    states.append(2**i)
    steps=(2**(i-1))-1
    nn_gate_count.append(nn_gate_count_cal(i))
    as_gate_count.append((((2**i)*4)+2)*steps)

plt.plot(states,nn_gate_count,c="green")
plt.plot(states,as_gate_count,c="red")
plt.xlabel("No. Of States")
plt.ylabel("No. Of Gates")
plt.title("Nearest Neighbour vs Adder Subtractor\nGreen=NN | Red=Adder Subtractor")
plt.show()
print("It Takes Eponentially(O(2^N)) Less Gate In Nearest Neighbour Approach Compare To Adder Subtractor Approach For 2^N States, Where N Is No. Of Bits")
```

```python
def nn_gate_depth_cal(n):
    step=(2**(n-1))-1
    s=0
    for i in range(step):
        if i%2 == 0:
            s+=4
        else:
            s+=6
    return s

s=6
states=[]
nn_gate_depth=[]
as_gate_depth=[]
for i in range(1,s):
    states.append(2**i)
    steps=(2**(i-1))-1
    nn_gate_depth.append(nn_gate_depth_cal(i))
    as_gate_depth.append((((2**i)*4)+2)*steps)

plt.plot(states,nn_gate_depth,c="green")
plt.plot(states,as_gate_depth,c="red")
plt.xlabel("No. Of Steps")
plt.ylabel("No. Of Depth")
plt.title("Nearest Neighbour vs Adder Subtractor\nGreen=NN | Red=Adder Subtractor")
plt.show()
print("It Has Eponentially(O(2^N)) Less Gate Depth In Nearest Neighbour Approach Compare to Adder Subtractor Approach For 2^N States, Where N Is The No. Of Bits")
```
